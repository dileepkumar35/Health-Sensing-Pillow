```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import csv
import random

# Open the CSV file for writing
with open('smart_pillow_data1.csv', 'w', newline='') as f:
    writer = csv.writer(f)

    # Write the header row
    header = ['Right', 'Middle', 'Left', 'Date', 'Time']
    writer.writerow(header)
    
    # Generate data for one day
    for day in range(1, 32):
        for hour in range(24):
            # Define the patterns
            patterns = [[1, 0, 0], [0, 1, 0], [0, 0, 1]]
            random.shuffle(patterns)  # Shuffle the patterns for randomness

            # Repeat each pattern at least 10 times
            for pattern in patterns:
                for _ in range(random.randint(10, 15)):  # Repeat pattern for at least 10 rows
                    right, middle, left = pattern

                    date = f"{day:02d}-May-2024"  # Assuming May 2024 for simplicity
                    time = f"{hour:02d}:{random.randint(0, 59):02d}:00"

                    # Write the data row
                    row = [right, middle, left, date, time]
                    writer.writerow(row)

print("CSV file generated successfully!")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import time
import shutil
while True:
    df=pd.read_csv('smart_pillow_data1.csv')

    # Melt the DataFrame to long format
    df_melted = pd.melt(df, id_vars=['Date', 'Time'], var_name='Pattern', value_name='Active')
    
    # Plotting
    plt.figure(figsize=(12, 6))
    
    # Create count plot for Right pattern
    plt.subplot(1, 3, 1)
    sns.countplot(data=df_melted[df_melted['Pattern'] == 'Right'], x='Active', palette='Set2')
    plt.title('Right Pattern')
    plt.xlabel('Activity')
    plt.ylabel('Frequency')
    
    # Create count plot for Center pattern
    plt.subplot(1, 3, 2)
    sns.countplot(data=df_melted[df_melted['Pattern'] == 'Middle'], x='Active', palette='Set2')
    plt.title('Middle Pattern')
    plt.xlabel('Activity')
    plt.ylabel('Frequency')
    
    # Create count plot for Left pattern
    plt.subplot(1, 3, 3)
    sns.countplot(data=df_melted[df_melted['Pattern'] == 'Left'], x='Active', palette='Set2')
    plt.title('Left Pattern')
    plt.xlabel('Activity')
    plt.ylabel('Frequency')
    
    plt.tight_layout()
    plt.savefig('patten.png')
    shutil.copy("patten.png","D:/smart pillow/Pillow_Server/web/p.png")
    
    # Show plots
    plt.show()
    # Count the occurrences of each pattern
    pattern_counts = df[['Right', 'Middle', 'Left']].sum()
    
    # Plotting
    plt.figure(figsize=(8, 6))
    
    # Plot bar chart
    pattern_counts.plot(kind='bar', color=['blue', 'orange', 'green'])
    
    plt.title('Frequency of Patterns')
    plt.xlabel('Pattern')
    plt.ylabel('Frequency')
    plt.xticks(rotation=0)
    plt.tight_layout()
    plt.savefig('Frequency.png')
    shutil.copy("Frequency.png","D:/smart pillow/Pillow_Server/web/f.png")
    # Show plot
    plt.show()

    time.sleep(10)
```
